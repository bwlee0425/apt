```python
import requests

class TransactionPrice:
    def __init__(self, service_key=None):
        self.service_key = service_key
        self.meta_dict = {
            "아파트": {
                "매매": {
                    "url": "http://apis.data.go.kr/1613000/RTMSDataSvcAptTradeDev/getRTMSDataSvcAptTradeDev",
                },
                "전월세": {
                    "url": "http://apis.data.go.kr/1613000/RTMSDataSvcAptRent/getRTMSDataSvcAptRent",
                },
            },
            # 다른 부동산 유형도 여기에 추가
        }

    def check_api_services(self):
        for property_type, trade_types in self.meta_dict.items():
            for trade_type, info in trade_types.items():
                url = info["url"]
                params = {
                    "serviceKey": self.service_key,
                    "pageNo": 1,
                    "numOfRows": 1,
                    "LAWD_CD": "11110",  # 예시 시군구 코드
                    "DEAL_YMD": "202301"  # 예시 연월
                }
                try:
                    response = requests.get(url, params=params)
                    if response.status_code == 200:
                        print(f"{property_type} {trade_type} API 요청 성공!")
                        print("응답 데이터:", response.json())
                    else:
                        print(f"{property_type} {trade_type} API 요청 실패: 상태 코드 {response.status_code}")
                except Exception as e:
                    print(f"{property_type} {trade_type} API 요청 중 오류 발생:", e)

# 서비스 키 설정
serviceKey = "uzKxKGshXjPznI62iA3GM9P44gSwHpX9sb3qprkoQrK0Nrzp7H6sy31XxrZb1%2BAV9W21EhRX%2Bdq0%2BQ4J3hXazA%3D%3D"

# TransactionPrice 인스턴스 생성
ts = TransactionPrice(serviceKey)

# API 서비스 확인
ts.check_api_services()
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
import time

# 엑셀 파일에서 데이터 읽기
data3 = pd.read_excel('result_202410.xlsx')  # 파일 이름을 적절히 수정하세요.
geolocator = Nominatim(user_agent="your_unique_name")

# 결과를 저장할 리스트
latitudes = []
longitudes = []
status = []  # 처리 상태를 저장할 리스트

# 최대 재시도 횟수
max_retries = 1

# 주소 처리
for index, row in data3.iterrows():
    address = f"{row['umdNm']} {row['roadNm']} {row['bonbun']}"
    print(f"Processing address: {address}")
    
    retries = 0
    success = False
    
    while retries < max_retries:
        try:
            location = geolocator.geocode(address, timeout=10)
            if location:
                latitudes.append(location.latitude)
                longitudes.append(location.longitude)
                status.append("완료")  # 처리 완료
                print(f"Success: {address} -> ({location.latitude}, {location.longitude})")
                success = True
                break  # 성공하면 루프 종료
            else:
                latitudes.append(None)
                longitudes.append(None)
                status.append("실패")  # 처리 실패
                print(f"Failed to geocode: {address}")
                break  # 실패하면 루프 종료
        except GeocoderTimedOut:
            print("Geocoding timed out, retrying...")
            retries += 1
            time.sleep(1)  # 잠시 대기 후 재시도
        except Exception as e:
            print(f"Error processing {address}: {e}")
            latitudes.append(None)
            longitudes.append(None)
            status.append("실패")  # 처리 실패
            break  # 예외 발생 시 루프 종료

    if not success:  # 모든 재시도 실패 시
        latitudes.append(None)
        longitudes.append(None)
        status.append("실패")  # 처리 실패

# 리스트 길이 확인
print(f"Length of latitudes: {len(latitudes)}")
print(f"Length of longitudes: {len(longitudes)}")
print(f"Length of status: {len(status)}")
print(f"Length of data3: {len(data3)}")

# 결과를 데이터프레임에 추가
# 데이터프레임의 길이에 맞춰 None 추가하기
while len(latitudes) < len(data3):
    latitudes.append(None)
while len(longitudes) < len(data3):
    longitudes.append(None)
while len(status) < len(data3):
    status.append("실패")

# 데이터프레임에 추가
data3['위도'] = latitudes[:len(data3)]
data3['경도'] = longitudes[:len(data3)]
data3['처리 상태'] = status[:len(data3)]

# 처리 완료 및 실패 개수 계산
completed_count = status.count("완료")
failed_count = status.count("실패")

# 결과 출력
print(f"총 주소 개수: {len(data3)}")
print(f"처리 완료: {completed_count}개")
print(f"처리 실패: {failed_count}개")

# 결과를 엑셀 파일로 저장
data3.to_excel('처리결과.xlsx', index=False)  # 결과 파일 이름을 적절히 수정하세요.
```
